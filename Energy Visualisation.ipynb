# Visualising the data

This is a visualisation of some of the data about energy sources in the UK from the Our World in Data dataset.

```python
# The usual imports

import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Ben Fairbairn\owid-energy-data.csv")
```

```python
df.head()
```

```python
df_UK = df[df['country'] == 'United Kingdom']
df_UK.head()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df_UK['year'], df_UK['gas_share_elec'], label='gas', c='blue')
plt.plot(df_UK['year'], df_UK['solar_share_elec'], label='solar', c='yellow')
plt.plot(df_UK['year'], df_UK['wind_share_elec'], label='wind', c='red')
plt.plot(df_UK['year'], df_UK['nuclear_share_elec'], label='nuclear', c='green')
plt.plot(df_UK['year'], df_UK['coal_share_elec'], label='coal', c='black')
plt.plot(df_UK['year'], df_UK['oil_share_elec'], label='oil', c='grey')
plt.plot(df_UK['year'], df_UK['biofuel_share_elec'], label='biofuel', c='brown')
plt.plot(df_UK['year'], df_UK['hydro_share_elec'], label='hydro', c='purple')

plt.legend()
plt.title('UK electricity sources 1985-2022')
plt.ylabel('%')
plt.xlabel('year')
#plt.xlim([1900,2000])

plt.show()
```
